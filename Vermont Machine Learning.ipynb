```python
from sklearn.linear_model import LinearRegression,Lasso, Ridge
from sklearn import metrics
from math import sqrt
```

```python
lm = LinearRegression()
lm.fit(X_dummy_train, y_train)
y_predict = lm.predict(X_dummy_test)
print(sqrt(metrics.mean_squared_error(y_test, y_predict)))
```

```python
# Lasso Regression
lm_lasso = Lasso(alpha=0.1)
lm_lasso.fit(X_dummy_train, y_train)
y_predict = lm_lasso.predict(X_dummy_test)
print(sqrt(metrics.mean_squared_error(y_test, y_predict)))
print(np.count_nonzero(lm_lasso.coef_))
```

```python
# Ridge Regression
lm_ridge = Ridge(alpha=0.1, normalize=True)
lm_ridge.fit(X_dummy_train, y_train)
y_predict = lm_ridge.predict(X_dummy_test)
print(sqrt(metrics.mean_squared_error(y_test, y_predict)))
```

```python
# Decision Trees

# Instantiate a DecisionTreeRegressor (with random_state=1).
from sklearn.tree import DecisionTreeRegressor

treereg = DecisionTreeRegressor(random_state=1)
treereg
```

```python
# Use leave-one-out cross-validation (LOOCV) to estimate the RMSE for this model.
from sklearn.model_selection import cross_val_score
scores = cross_val_score(treereg, X_dummy_train, y_train, cv=14, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores))
```

```python
# Try different values one by one.
treereg = DecisionTreeRegressor(max_depth=1, random_state=1)
scores = cross_val_score(treereg, X_dummy_train, y_train, cv=14, scoring='neg_mean_squared_error')
np.mean(np.sqrt(-scores))
```

```python
# List of values to try:
max_depth_range = list(range(1, 12))

# List to store the average RMSE for each value of max_depth:
RMSE_scores = []
MSE_trainset_score = []

# Use LOOCV with each value of max_depth.
for depth in max_depth_range:
    treereg = DecisionTreeRegressor(max_depth=depth, random_state=1)
    MSE_scores = cross_val_score(treereg, X_dummy_train, y_train, cv=14, scoring='neg_mean_squared_error')
    treereg.fit(X_dummy_train, y_train)
    y_predict = treereg.predict(X_dummy_train)
    MSE_trainset_score.append(sqrt(metrics.mean_squared_error(y_train, y_predict)))
    RMSE_scores.append(np.mean(np.sqrt(-MSE_scores)))
```

```python
# Plot max_depth (x-axis) versus RMSE (y-axis).
plt.plot(max_depth_range, RMSE_scores);
plt.xlabel('max_depth');
plt.ylabel('RMSE (lower is better)');
```

```python
# Plot max_depth (x-axis) versus RMSE (y-axis).
plt.plot(max_depth_range, MSE_trainset_score);
plt.xlabel('max_depth');
plt.ylabel('RMSE (lower is better)');
```
